```python

import unittest
import demo
from os import listdir
from os.path import isfile, join
import collections
```

```python
files_path = './inputs/' #synthetic_networks/data/
onlyfiles = [f for f in listdir(files_path) if isfile(join(files_path, f))]
# print(onlyfiles)
```

```python
# For no attributes data, I am still adding lables for items in facts.. need to check if this is 
# corrrect if it is better to remove them completely.

def processSyntheticPlantedPartition():

    def isclose(a, b, rel_tol=1e-09, abs_tol=0.0):
        return abs(a-b) <= max(rel_tol * max(abs(a), abs(b)), abs_tol)


#     files_path = './inputs/' #synthetic_networks/data/
#     onlyfiles = [f for f in listdir(files_path) if isfile(join(files_path, f))]

    sorted(onlyfiles)   

    probabs_initial_accuracies = {}
    probabs_final_accuracies = {}
    probabs_c_in_minus_c_outs = {}

    for prob in range(4, 10, 1):
        attribute_match_prob = 0.1 * prob

        initial_accuracies = []
        final_accuracies = []
        c_ins = []
        c_outs = []
        c_in_minus_c_outs = []
        n =0

#         print(facts_file)
        #planted_partition_with_attributes_11.0_5.0.cfacts
        for facts_file in onlyfiles:
            if 'planted_partition_with_attributes_1000' in facts_file and 'cfacts' in facts_file:


                # planted_partition_with_attributes_11.0_5.0.cfacts

                file_name = facts_file.split('.cfacts')[0]
                file_prob = float(file_name.split('_')[-1])

#                 print(file_name)
                if isclose(file_prob, attribute_match_prob):
                    print(facts_file)
                    
                    n = int(file_name.split('_')[4])
                    c_in = float(file_name.split('_')[6])
                    c_out = float(file_name.split('_')[7])
                    # planted_partition_with_attributes_12.0_4.0
                    init_acc,final_acc = demo.runMain(("--stem "+file_name).split())

                    initial_accuracies.append(init_acc)
                    final_accuracies.append(final_acc)

                    c_ins.append(c_in)
                    c_outs.append(c_out)
                    c_in_minus_c_outs.append(c_in-c_out )

        print('c_ins', c_ins)
        print('c_outs', c_outs) 
        
        print('c_in_minus_c_outs' , c_in_minus_c_outs)   
        print('initial_accuracies' , initial_accuracies)
        print('final_accuracies' , final_accuracies)
        
        probabs_initial_accuracies[str(attribute_match_prob)] = initial_accuracies
        probabs_final_accuracies[str(attribute_match_prob)] = final_accuracies
        probabs_c_in_minus_c_outs[str(attribute_match_prob)] = c_in_minus_c_outs
        
    return probabs_c_in_minus_c_outs, probabs_initial_accuracies, probabs_final_accuracies
        
        
probabs_c_in_minus_c_outs, probabs_initial_accuracies, probabs_final_accuracies = processSyntheticPlantedPartition()
```

```python
print(probabs_initial_accuracies.keys())
print(collections.OrderedDict(sorted(probabs_initial_accuracies.items())).keys())

print(probabs_initial_accuracies.values())
print(probabs_c_in_minus_c_outs.keys())
```

```python
import pylab

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 26}

pylab.rc('font', **font)


def plot_results(probabs_c_in_minus_c_outs, probabs_accuracies, y_label, save_name_file= ''):
    pylab.figure(figsize=(18,12))

    sorted_probabs_accuracies = collections.OrderedDict(sorted(probabs_accuracies.items(), reverse=True))

    legends_label = {}

    for key in sorted_probabs_accuracies.keys():
            if float(key) >= 0.5:
                legends_label[key] = 'Correct Attribute Prob: ' + key
            else:
                legends_label[key] = 'Without Attributes'

    lines = ["--r^", "--g^", "-b+","-r+","-.g*" ,"-.b*",":ro",":go"]


    i = 0
    for probab in sorted_probabs_accuracies.keys():

        accuracies = probabs_accuracies[probab]
        c_in_minus_c_out= probabs_c_in_minus_c_outs[probab]

        c_in_minus_c_out_accuracies= {}
        for j in range(0, len(c_in_minus_c_out)):
            c_in_minus_c_out_accuracies[c_in_minus_c_out[j]] = accuracies[j]


        # need to sort accuracies as per c_in_minus_c_out
        ordered_c_in_minus_c_out_accuracies = collections.OrderedDict(sorted(c_in_minus_c_out_accuracies.items()))

        pylab.plot(ordered_c_in_minus_c_out_accuracies.keys(),\
                   ordered_c_in_minus_c_out_accuracies.values(), lines[i], linewidth=2, label= legends_label[probab])

        pylab.legend(bbox_to_anchor=(0.49, 0.96), loc=1, borderaxespad=0)
        pylab.ylabel(y_label, fontsize= 40)
        pylab.xlabel('Centrality_in - Centrality_out' , fontsize=40)

        i = i +1
    #     pylab.title(emb_file)

    # pylab.ylim(-1.5, 2.0)
    pylab.show()
#     pylab.savefig('./inputs/synthetic_networks/plots/' +y_label  + save_name_file + '.png')
    
plot_results(probabs_c_in_minus_c_outs, probabs_initial_accuracies, 'Initial Accuracy', '_with_and_without_attributes')
```

```python
plot_results(probabs_c_in_minus_c_outs, probabs_final_accuracies, 'Final Accuracy', '_with_and_without_attributes')
```
